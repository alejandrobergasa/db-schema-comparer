```python
import re
```

```python
with open("des.txt", "r") as f:
    file1 = f.read()
    
file1 = file1.split("\n")

bad_cols = (
    ["Nombre","Â¿Nulo?","Tipo"],
    )

file1 = [f for f in file1 if f.split() not in bad_cols and not f.replace(" ","").startswith("-----")]

tables1 = {}
last_table = ""

for l in file1:
    if l.startswith("{Description of table "):
        table_name = l[22:l.find("}")]
        tables1.update({table_name: {"COLUMNS": [], "TYPES": [], "NULLS": [],}})
        last_table = table_name
    else:
        col_name = l[:l.find(" ")]
        if " NOT NULL " in l:
            null = False
            col_type = l[l.find(" NOT NULL ")+10:].lstrip().rstrip()
        else:
            null = True
            col_type = l[l.find(" "):].lstrip().rstrip()
        tables1[table_name]["COLUMNS"].append(col_name)
        tables1[table_name]["TYPES"].append(col_type)
        tables1[table_name]["NULLS"].append(null)

tables1
```
